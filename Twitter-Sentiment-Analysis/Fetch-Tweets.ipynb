```python
bearer_token=config.ACADEMIC_BEARER
```

```python
def create_headers(bearer_token):
    headers = {"Authorization": "Bearer {}".format(bearer_token)}
    return headers


def connect_to_endpoint(url, headers, params):
    response = requests.request(
        "GET", searchUrl, headers=headers, params=params)
    print(response.status_code)
    if response.status_code != 200:
        raise Exception(response.status_code, response.text)
    return response.json()
```

```python
# Optional params: start_time,end_time,since_id,until_id,max_results,next_token,
# expansions,tweet.fields,media.fields,poll.fields,place.fields,user.fields
query_params2 = {'query': 'WorldCup OR Qatar22 OR worldcup OR (World Cup) place_country:EG lang:en has:geo -is:retweet',
                 'start_time': '2022-12-18T00:00:00Z',
                 'end_time': '2022-12-19T00:00:00Z',
                'tweet.fields': 'created_at,public_metrics,geo',
                'expansions': 'author_id,geo.place_id',
                'user.fields': 'location',
                'place.fields': 'country,country_code',
                'max_results': 500
                }
```

```python
def main():
    with open('DecTweets/WCTweets.json', 'a') as outfile:
        headers = create_headers(bearer_token)
        json_response = connect_to_endpoint(searchUrl, headers, query_params2)
        tweet_count = json_response['meta']['result_count']
        print("The number of tweets retrieve is: ", tweet_count)
        next_token = json_response['meta'].get('next_token', None)
        print("The next_token is: ", next_token)
        json.dump(json_response, outfile)
        num = 1

        while (next_token != None):
            num += 1
            filename = 'DecTweets/WCTweets'+str(num)+'.json'
            with open(filename, 'a') as outfile:
                query_params_new = {'query': 'WorldCup OR Qatar22 OR worldcup OR (World Cup) has:geo lang:en -is:retweet',
                                    'start_time': '2022-12-18T00:00:00Z',
                                    'end_time': '2022-12-19T00:00:00Z',
                                    'next_token': next_token,
                                    'tweet.fields': 'created_at,public_metrics,geo',
                                    'expansions': 'author_id,geo.place_id',
                                    'user.fields': 'location',
                                    'place.fields': 'country,country_code',
                                    'max_results': 500
                                    }
                json_response = connect_to_endpoint(
                    searchUrl, headers, query_params_new)
                next_token = json_response['meta'].get('next_token', None)
                print("The next_token is: ", next_token)
                tweet_count += json_response['meta']['result_count']
                print("The accumulative number of tweets retrieve is: ", tweet_count)
                json.dump(json_response, outfile)
                time.sleep(1)

if __name__ == "__main__":
    main()
```
